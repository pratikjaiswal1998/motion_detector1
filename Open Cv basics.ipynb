```python
import cv2,time,datetime
import pandas
first_frame = None
status_list = [None,None]
times = []
df = pandas.DataFrame(columns = ['Start','End'])#dataframe to store start and end of object
video = cv2.VideoCapture(0)

while True:#loop chaldo for frames
    
    check, frame = video.read()
    status=0
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)#color b&w krdo
    gray = cv2.GaussianBlur(gray,(21,21),0)#gaussian blur me convert kro
    if first_frame is None:#pehla frame ko store
        first_frame = gray
        continue
        
    delta_frame = cv2.absdiff(first_frame,gray)#diff calculate
    thresh_delta = cv2.threshold(delta_frame,30,255,cv2.THRESH_BINARY)[1]#2 lines diff less thn 30 to black,else white pixels
    thresh_delta = cv2.dilate(thresh_delta,None,iterations = 0)#
    #define contour:
    (_,cnts,_) = cv2.findContours(thresh_delta.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    #remove noise such as house fly
    for contour in cnts:#keep white having area >1000
        if cv2.contourArea(contour) < 1000:
            continue
        status = 1#object detect hua sttus change
        (x,y,w,h) = cv2.boundingRect(contour)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)#object ke around box
    
    status_list.append(status)#list of status for every frame
    status_list = status_list[-2:]
    #record determine kro jab bhi change ho
    if status_list[-1] == 1 and status_list[-2] ==0:
        times.append(datetime.datetime.now())
    if status_list[-1] == 0 and status_list[-2] == 1:
        times.append(datetime.datetime.now())
    
    cv2.imshow('frame',frame)
    cv2.imshow('Capturing',gray)
    cv2.imshow('delta',delta_frame)
    cv2.imshow('thresh',thresh_delta)
    key = cv2.waitKey(1)#window opening time(generate new frame after every milli second)
    if key == ord('q'):#q dabake window close
        break

print(status_list)
print(times)

for i in range(0,len(times),2):#store time values in df
    df = df.append({'Start':times[i],'End':times[i+1]},ignore_index = True)
df.to_csv("Times.csv")

video.release()
cv2.destroyAllWindows()


```
